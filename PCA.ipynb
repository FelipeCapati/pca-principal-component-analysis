```python
import numpy as np
from pca import PCA, PcaType
from lms import LMS, LmsType
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Exercise 1 - Alps Water Dataset

```python
data = np.array([
        [194.5, 194.3, 197.9, 198.4, 199.4, 199.9, 200.9, 201.1, 201.4, 201.3, 203.6, 204.6, 209.5, 208.6, 210.7, 211.9, 212.2],
        [20.79, 20.79, 22.40, 22.67, 23.15, 23.35, 23.89, 23.99, 24.02, 24.01, 25.14, 26.57, 28.49, 27.76, 29.04, 29.88, 30.06]
])
```

```python
X = [
[194.5],
[194.3],
[197.9],
[198.4],
[199.4],
[199.9],
[200.9],
[201.1],
[201.4],
[201.3],
[203.6],
[204.6],
[209.5],
[208.6],
[210.7],
[211.9],
[212.2]
]

Y = [
[20.79],
[20.79],
[22.40],
[22.67],
[23.15],
[23.35],
[23.89],
[23.99],
[24.02],
[24.01],
[25.14],
[26.57],
[28.49],
[27.76],
[29.04],
[29.88],
[30.06]
]
```

```python
pca = PCA()
pca.fit(data=data, pcaType=PcaType.Two)
data_pca = pca.transform(data=data, n_components=2)
```

```python
lms = LMS()

lms.fit(x=X, y=Y, type=LmsType.Linear)
y_predict_linear = lms.predict(x=X)
```

```python
data_plot_x = data[0] - np.average(data[0])
data_plot_y = data[1] - np.average(data[1])
data_pca_plot_x = data_pca[0] - np.average(data_pca[0])
data_pca_plot_y = data_pca[1] - np.average(data_pca[1])
np_x = np.array([(X[i][0]) for i in range(0, len(X))])
np_x_plot = np_x - np.average(np_x)
np_y = np.array([(y_predict_linear[i][0]) for i in range(0, len(y_predict_linear))])
np_y_plot = np_y - np.average(np_y)
plt.scatter(data_plot_x, data_plot_y, color="black", label="Data")
plt.plot(data_pca_plot_x, data_pca_plot_y, color="brown", label="PCA - Transform Data")
plt.plot(np_x_plot, np_y_plot, color="blue", label="MMQ - Linear Predict")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Alps Water Dataset')
plt.legend()
plt.grid()
```
